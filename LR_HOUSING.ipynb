```
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt


dataset = datasets.fetch_california_housing()

data_x = dataset.data[:]

data_y = dataset.target[:]

x_train = data_x[:-20]

x_test = data_x[-20:]

y_train = data_y[:-20]

y_test = data_y[-20:]

model = LinearRegression()

model.fit(x_train,y_train)

y_pred = model.predict(x_test)


# The coefficients
print('\n Coefficients: \n', model.coef_)

# The Intercept
print('\n Intercept: \n', model.intercept_)

# The mean squared error
print("\n Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))

# Explained variance score: 1 is perfect prediction
# R^2 (coefficient of determination) regression score function
print('\n Variance score: %.2f' % r2_score(y_test, y_pred))

print(x_test[:,2].shape)

print(y_pred.shape)


plt.scatter(x_test[:,2], y_test,  color='black')
plt.plot(x_test[:,2], y_pred, color='blue', linewidth=3)

plt.show()

```
